Explanation of each stage in the **Pipeline Architecture**:
1. **Data Ingestion**

* **Objective**: Load data from multiple CSV files (`coin_gecko_2022-03-16.csv` and `coin_gecko_2022-03-17.csv`).
* **Tools**: `pandas`, `glob`, `os`
* **Actions**:

  * Combine data from multiple days
  * Handle encoding and missing values
  * Standardize date/time formats

---
2. **Data Preprocessing**

* **Objective**: Clean and prepare raw data for analysis.
* **Actions**:

  * Handle null or duplicate rows
  * Convert categorical data
  * Normalize/standardize numerical features
  * Ensure consistent column formats

---
3. **Exploratory Data Analysis (EDA)**

* **Objective**: Understand data distributions and relationships.
* **Tools**: `matplotlib`, `seaborn`, `pandas_profiling`, `plotly`
* **Actions**:

  * Visualize top coins by liquidity and volume
  * Correlation analysis
  * Detect outliers, trends, and patterns

---
4. **Feature Engineering**

* **Objective**: Create new features to improve model performance.
* **Actions**:

  * Calculate liquidity ratios (e.g., `volume_24h / market_cap`)
  * Create trend indicators
  * One-hot encode categories
  * Rank coins by liquidity percentile

---
5. **Model Training**

* **Objective**: Train ML models to predict or rank coins based on liquidity.
* **Models**: `Logistic Regression`, `Random Forest`, `XGBoost`, `SVM`
* **Actions**:

  * Train/test split
  * Apply scaling
  * Fit models and evaluate with metrics (e.g., Accuracy, F1-score)

---
6. **Model Output**

* **Objective**: Produce actionable outputs (e.g., ranked list of coins).
* **Actions**:

  * Display top N coins by predicted liquidity
  * Save results to CSV or dashboard
  * Optionally deploy as an API or stream

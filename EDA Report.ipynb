```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data to define the DataFrame
data = {
	"coin": ["Bitcoin", "Ethereum", "Tether", "Shiba Inu", "THORChain"],
	"symbol": ["BTC", "ETH", "USDT", "SHIB", "RUNE"],
	"price": [40851, 2824, 1, 0.00002235, 7.23],
	"1h": [0.01, 0.029, 0, -0.002, -5.9],
	"24h": [0.01, 0.029, 0, -0.002, -5.9],
	"7d": [5.5, 2.9, 0, -19.9, -55],
	"24h_volume": [20480000000, 13640000000, 44130000000, 1000000, 5000000],
	"mkt_cap": [776080000000, 330000000000, 80000000000, 12300000000, 2000000000],
}

# Create the DataFrame
df = pd.DataFrame(data)

# Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["price"], bins=50, kde=True)
plt.xscale("log")  
plt.title("Cryptocurrency Price Distribution")
plt.xlabel("Price (Log Scale)")
plt.ylabel("Frequency")
plt.show()

# Market Cap vs. 24h Volume (Liquidity Analysis)
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["mkt_cap"], y=df["24h_volume"], hue=df["coin"], alpha=0.7)
plt.xscale("log")
plt.yscale("log")
plt.title("Market Cap vs 24h Trading Volume")
plt.xlabel("Market Cap ($, Log Scale)")
plt.ylabel("24h Trading Volume ($, Log Scale)")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Crypto Metrics")
plt.show()
```
